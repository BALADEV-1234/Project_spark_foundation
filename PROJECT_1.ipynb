#### train test spliting of data 

```python
X = s_data.iloc[:, :-1].values  
y = s_data.iloc[:, 1].values  
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0) 
```

```python
#Statistically showing the train and test data are belongs to Population dataset.
```

```python
#Ho:The Sample belongs to Population.
#Ha:The Sample doesnot belongs to Population
static,pval=ttest_ind(X_train,X_test)
if pval>0.05:
    print('Fail to reject null hypothesis.The Sample belongs to Population.')
else:
    print('reject null hypothesis.The Sample doesnot  belongs to Population.')
```

### MODEL BUILDING

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 
```

```python
# Plotting the regression line
# Plotting for the test data
fig = px.scatter(s_data, x='Hours', y='Scores',opacity=0.58,trendline='ols', trendline_color_override='darkblue')
fig.show()
```

```python
y_pred = regressor.predict(X_test) # Predicting the scores
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
# You can also test with your own data
hours = 9.25
own_pred = regressor.predict(np.array(hours).reshape(-1,1))
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

### Evaluating the model

```python
print('Mean Absolute Error of the test data:',mean_absolute_error(y_test, y_pred)) 
print('Mean Absolute Error of the train data:',mean_absolute_error(y_train,regressor.predict(X_train))) 
```

```python
print('RMSE of the test data:',np.sqrt(mean_squared_error(y_test, y_pred)))
print('RMSE of the train data:',np.sqrt(mean_squared_error(y_train,regressor.predict(X_train))))
```
